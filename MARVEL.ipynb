```python
# 1. Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# 2. Read csv
df = pd.read_csv("/content/charcters.csv")
```

```python
# 3. Show first records from csv
print(df.head())
```

```python
# 4. Show number of rows and columns
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
# 5. You need to find the values of alignment ,can use value_counts()
print(df['Alignment'].value_counts())
```

```python
# 6. Find out only good alignment holders superheroes
good_superheroes = df[(df['Alignment'] == 'good')] 
```

```python
# 7. Show first five records which you found in point 6
print(good_superheroes.head())
```

```python
# 8. Show top five records having top speed of heroes of good alignment
top_speed_good_heroes = good_superheroes.sort_values(by=['Speed'], ascending=False).head()
print(top_speed_good_heroes)
```

```python
# 9. Show 5 records of super heroes who have maximum power of good alignment
max_power_good_heroes = good_superheroes.sort_values(by=['Power'], ascending=False).head()
print(max_power_good_heroes)
```

```python
# 10. Find out how many super heroes are there with power 100 of good alignment
count_max_power_100 = good_superheroes[good_superheroes['Power'] == 100].shape[0]
print("Number of super heroes with power 100:", count_max_power_100)
```

```python
# 11. Shape them what you got in point 10
shape_max_power_100 = max_power_good_heroes[max_power_good_heroes['Power'] == 100].shape
print("Shape of super heroes with power 100:", shape_max_power_100)
```

```python
# 12. Show all records from point 10
max_power_good_heroes = good_superheroes.sort_values(by=['Power'], ascending=False)
max_power_100_good_heroes = max_power_good_heroes[max_power_good_heroes['Power'] == 100]
print(max_power_100_good_heroes)
```

```python
# 13. Retrieve total of first five records of max power of good alignment super heroes
total_top_max_power_good_heroes = max_power_good_heroes.head()['Total'].sum()
print("Total of first five records of max power of good alignment super heroes:", total_top_max_power_good_heroes)
```

```python
# 14. Draw a bar plot of all super heroes who are having good alignment and max power of top five only , take same object of point 13 , show name and total in plot with green bars
import matplotlib.pyplot as plt
max_power_100_good_heroes1= max_power_100_good_heroes.head()
plt.bar(max_power_100_good_heroes1['Name'], max_power_100_good_heroes1['Total'], color='green')
plt.xlabel("Name")
plt.xticks(rotation=45)
plt.ylabel("Total")
plt.title("Super Heroes with Good Alignment and Max Power of 100")
plt.show()
```

```python
# 15. Extract villains having bad alignment
bad_villains = df[(df['Alignment'] == 'bad')] 
```

```python
# 16. Show first five records of point 15
print(bad_villains.head())
```

```python
# 17. Show top five fastest super villains in terms of super speed
top_speed_bad_villains = bad_villains.sort_values(by=['Speed'], ascending=False).head()
print(top_speed_bad_villains)
```

```python
# 18. Top five super villains in terms of intelligence
top_intelligence_villains = bad_villains.sort_values(by=['Intelligence'], ascending=False).head()
print(top_intelligence_villains)
```

```python
# 19. Show who is most dangerous super villain after calculating their total (top 5 only)
top_total_bad_villains = bad_villains.sort_values(by=['Total'], ascending=False).head()

most_dangerous_villain = top_total_bad_villains.iloc[0]['Name']
print("The most dangerous super villain is:", most_dangerous_villain)
```

```python
# 20. Draw a histogram for speed of super heroes having fig size 10,5, provide speed in histogram for only good alignment super heroes, title should be "distribution of speed", xlabel should be "speed"
import matplotlib.pyplot as plt

# Filter the dataframe for only good alignment super heroes
good_superheroes = df.loc[df['Alignment'] == 'good']

# Create a histogram for the speed column of the good_superheroes dataframe
plt.figure(figsize=(10, 5))
plt.hist(good_superheroes['Speed'], bins=20)
plt.title('Distribution of Speed')
plt.xlabel('Speed')
plt.ylabel('Count')
plt.show()
```

```python
# 21. Draw a histogram for combat of super villains having fig size 10,5, provide combat in histogram for only bad alignment super heroes, title should be "distribution of combat", xlabel should be "combat"
import matplotlib.pyplot as plt

# Filter the dataframe for only bad alignment super villains
bad_villains = df.loc[df['Alignment'] == 'bad']

# Create a histogram for the combat column of the bad_villains dataframe
plt.figure(figsize=(10, 5))
plt.hist(bad_villains['Combat'], bins=20)
plt.title('Distribution of Combat for Bad Alignment Super Villains')
plt.xlabel('Combat')
plt.ylabel('Count')
plt.show()
```

```python
# 22.Scatter plot showing the relationship between intelligence and combat for all characters
plt.scatter(df['Intelligence'], df['Combat'], alpha=0.5)
plt.title('Relationship between Intelligence and Combat')
plt.xlabel('Intelligence')
plt.ylabel('Combat')
plt.show()
```

```python
# 23.Histogram showing the distribution of power for all characters
plt.hist(df['Power'], bins=20)
plt.title('Distribution of Power for All Characters')
plt.xlabel('Power')
plt.ylabel('Count')
plt.show()
```

```python
#24. Two histograms showing the distribution of strength for good and bad alignment characters
good_chars = df.loc[df['Alignment'] == 'good']
bad_chars = df.loc[df['Alignment'] == 'bad']

plt.hist(good_chars['Strength'], alpha=0.5, label='Good Alignment', bins=20)
plt.hist(bad_chars['Strength'], alpha=0.5, label='Bad Alignment', bins=20)
plt.title('Distribution of Strength by Alignment')
plt.xlabel('Strength')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
#25. Correlation matrix showing the correlation between all attributes for all characters
corr = df[['Intelligence', 'Strength', 'Durability', 'Speed', 'Combat', 'Power', 'Total']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix for All Attributes')
plt.show()
```

```python
#26. Scatter plot showing the relationship between strength and durability with color-coded alignment
sns.scatterplot(x='Strength', y='Durability', hue='Alignment', data=df, alpha=0.5)
plt.title('Relationship between Strength and Durability with Color-Coded Alignment')
plt.xlabel('Strength')
plt.ylabel('Durability')
plt.show()
```

```python
#27. Pair plot showing the pairwise relationships between all attributes with color-coded alignment
sns.pairplot(data=df, vars=['Intelligence', 'Strength', 'Durability', 'Speed', 'Combat', 'Power'], hue='Alignment')
plt.suptitle('Pair Plot of All Attributes with Color-Coded Alignment')
plt.show()
```
